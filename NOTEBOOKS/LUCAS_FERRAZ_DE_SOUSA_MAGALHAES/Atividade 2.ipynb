```python
import pandas as pd


url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data'

column_names_pt_br = [
    'classe',
    'alcool',
    'acido_malico',
    'cinzas',
    'alcalinidade_de_cinzas',
    'magnesio',
    'fenois_totais',
    'flavanoides',
    'fenois_nao_flavanoides',
    'proantocianinas',
    'intensidade_de_cor',
    'matiz',
    'od280_od315_de_vinhos_diluidos',
    'prolina'
]


vinhos = pd.read_csv(url, names=column_names_pt_br, dtype={'classe': object})

contagem_classes = vinhos['classe'].value_counts()


estatisticas_descritivas = vinhos[['alcool', 'acido_malico', 'magnesio', 'matiz']].describe()

print("Contagem das classes:")
display(contagem_classes)

print("\nEstatísticas descritivas:")
display(estatisticas_descritivas)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


colunas_boxplot = ['magnesio', 'prolina', 'cinzas', 'alcool']

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))


for i, coluna in enumerate(colunas_boxplot):
    row = i // 2
    col = i % 2
    sns.boxplot(x=vinhos[coluna], ax=axes[row, col])
    axes[row, col].set_title(f'Boxplot de {coluna}')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
colunas_boxplot_classe = ['magnesio', 'fenois_totais', 'acido_malico', 'alcool']

for coluna in colunas_boxplot_classe:
  vinhos.boxplot(column=coluna, by='classe')
  plt.title(f'{coluna} por Classe')
  plt.suptitle('')
  plt.show()
```

```python

import matplotlib.pyplot as plt

plt.hist(vinhos['fenois_totais'], bins=10)
plt.xlabel('Fenois Totais')
plt.ylabel('Frequência')
plt.title('Fenois Totais')
plt.show()


vinhos_classe_3 = vinhos[vinhos['classe'] == '3']
plt.hist(vinhos_classe_3['fenois_totais'], bins=10)
plt.xlabel('Fenois Totais (Classe 3)')
plt.ylabel('Frequência')
plt.title('Fenois Totais (Classe 3)')
plt.show()
```
